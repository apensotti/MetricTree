```python
import pymysql
import pandas as pd

# Database connection
connection = pymysql.connect(
    host='127.0.0.1',
    user='root',
    password='Dodgetheblock@2',
    db='exactius'
)

# Load CSV data into DataFrame
df = pd.read_csv('C:/Users/alexp/src/MetricTree/metric_tree.csv', sep=',', header=0, encoding='utf-8')

df['first_revenue'].astype(float)
df['repeat_revenue'].astype(float)
df['contribution_margin'].astype(float)

# Renaming columns to the correct names
df.columns = [
    'date', 'year', 'month', 'week_start_date', 'week_number', 'week_number_year', 'trailing_7d', 'attribution_type',
    'channel_type', 'strategy', 'platform', 'channel', 'market', 'investment', 'agency_fees', 'violet_fee', 'sessions',
    'phone_calls', 'form_submissions', 'leads', 'first_appointments_completed', 'second_appointments_completed',
    'repeat_appointments_completed', 'first_revenue', 'repeat_revenue', 'twenty_one_days_revenue',
    'twenty_eight_days_revenue', 'ninety_days_revenue', 'one_hundred_eighty_days_revenue', 'contribution_margin'
]

with connection.cursor() as cursor:
        # Create table
        create_table_query = """
        CREATE TABLE IF NOT EXISTS metric_trees (
            date DATE,
            year INT,
            month VARCHAR(10),
            week_start_date DATE,
            week_number INT,
            week_number_year VARCHAR(10),
            trailing_7d VARCHAR(20),
            attribution_type VARCHAR(50),
            channel_type VARCHAR(50),
            strategy VARCHAR(50),
            platform VARCHAR(50),
            channel VARCHAR(50),
            market VARCHAR(50),
            investment DECIMAL(10,4),
            agency_fees DECIMAL(10,4),
            violet_fee DECIMAL(10,4),
            sessions DECIMAL(10,4),
            phone_calls DECIMAL(10,4),
            form_submissions DECIMAL(10,4),
            leads DECIMAL(10,4),
            first_appointments_completed DECIMAL(10,4),
            second_appointments_completed DECIMAL(10,4),
            repeat_appointments_completed DECIMAL(10,4),
            first_revenue DECIMAL(10,4),
            repeat_revenue DECIMAL(10,4),
            twenty_one_days_revenue DECIMAL(10,4),
            twenty_eight_days_revenue DECIMAL(10,4),
            ninety_days_revenue DECIMAL(10,4),
            one_hundred_eighty_days_revenue DECIMAL(10,4),
            contribution_margin DECIMAL(10,4)
        );
        """
        cursor.execute(create_table_query)
        connection.commit()


# Create insert query
insert_query = """
INSERT INTO metric_trees (
    date, year, month, week_start_date, week_number, week_number_year, trailing_7d, attribution_type,
    channel_type, strategy, platform, channel, market, investment, agency_fees, violet_fee, sessions,
    phone_calls, form_submissions, leads, first_appointments_completed, second_appointments_completed,
    repeat_appointments_completed, first_revenue, repeat_revenue, twenty_one_days_revenue,
    twenty_eight_days_revenue, ninety_days_revenue, one_hundred_eighty_days_revenue, contribution_margin
) VALUES (
    %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s
)
"""

# Insert data into MySQL table
with connection.cursor() as cursor:
    for _, row in df.iterrows():
        cursor.execute(insert_query, tuple(row))

# Commit the transaction
connection.commit()

# Close the connection
connection.close()
```
